## 3. Feature Engineering

### 3.1. some common and helpful features to extract:

**1. Lag Features**  
Capture recent history:

- **Close_t-1** â†’ yesterdayâ€™s price  
- **Close_t-2** â†’ price 2 days ago  
Useful for predicting today's or tomorrowâ€™s price

**2. Rolling Statistics**  
Capture trend and momentum:

- **RollingMean_7** â†’ 7-day average price  
- **RollingStd_7** â†’ 7-day standard deviation (volatility)  
Shows short-term trend and fluctuations

**3. Date/Time Features**  
Help detect seasonal patterns:

- **Day of Week** â†’ Monday, Tuesday, etc.  
- **Month** â†’ January, February, etc.  
- **Year, Quarter, Day**

**4. Returns**  
Calculate daily percentage change:

- **Return_1d** = (Today - Yesterday) / Yesterday  
Can help classify rising or falling days


```python
# Lag features
df_ts['Close_t-1'] = df_ts['Close'].shift(1)
df_ts['Close_t-2'] = df_ts['Close'].shift(2)

# Rolling statistics
df_ts['RollingMean_7'] = df_ts['Close'].rolling(window=7).mean()
df_ts['RollingStd_7'] = df_ts['Close'].rolling(window=7).std()

# Date-based features
df_ts['DayOfWeek'] = df_ts.index.dayofweek
df_ts['Month'] = df_ts.index.month
df_ts['Year'] = df_ts.index.year

# Daily return
df_ts['Return_1d'] = df_ts['Close'].pct_change()

# Display engineered features
df_ts.head(15)
```

**New Features Explained:**

| **Feature**       | **Description**                                   |
| ----------------- | ------------------------------------------------- |
| `Close_t-1`       | Yesterday's closing price                         |
| `Close_t-2`       | Closing price 2 days ago                          |
| `RollingMean_7`   | 7-day moving average (smooths trend)              |
| `RollingStd_7`    | 7-day rolling standard deviation (volatility)     |
| `DayOfWeek`       | Day of the week (0 = Monday, ..., 4 = Friday)     |
| `Month`           | Month number (1â€“12)                               |
| `Year`            | Year (for seasonal/yearly trends)                 |
| `Return_1d`       | Daily return = % change from previous day         |


These features are helpful for:

 **Time-series forecasting** (e.g., XGBoost, LSTM)

 **Signal classification** (e.g., buy/sell prediction)


### 3.2. Technical Stock Market Indicators

**1. RSI (Relative Strength Index)**  
RSI measures momentum â€” how strong recent price changes are. It tells you if a stock is **overbought** (too expensive) or **oversold** (too cheap).

**Range:**  
0 to 100  
- **Above 70** = Overbought â†’ might fall soon  
- **Below 30** = Oversold â†’ might rise soon

**Example:**  
- If **RSI = 80** â†’ price went up fast recently â†’ may correct downward  
- If **RSI = 20** â†’ price dropped a lot â†’ may bounce back up


**2. MACD (Moving Average Convergence Divergence)**  
MACD shows the relationship between two moving averages (fast and slow) to detect trend changes.

- **MACD Line** = 6-day EMA âˆ’ 14-day EMA  
- **Signal Line** = 4-day EMA of MACD  

We look for **crossovers** between MACD and Signal line.

**Signals:**  
- **MACD > Signal Line** â†’ Bullish (uptrend)  
- **MACD < Signal Line** â†’ Bearish (downtrend)  


**3. CMF (Chaikin Money Flow)**  
CMF combines price and volume to show whether money is flowing in (**buying**) or out (**selling**).

**Range:**  
Between -1 and +1  
- **Above 0** = buying pressure â†’ strength  
- **Below 0** = selling pressure â†’ weakness

**Example:**  
- If **CMF is +0.3**, it means buyers are more active.  
- If itâ€™s **-0.4**, sellers are dominating.


```python
import numpy as np

# RSI (Relative Strength Index)
def compute_rsi(series, period=14):
    delta = series.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df['RSI_14'] = compute_rsi(df['Close'], period=14)

# MACD (Moving Average Convergence Divergence)
ema_6 = df['Close'].ewm(span=6, adjust=False).mean()
ema_14 = df['Close'].ewm(span=14, adjust=False).mean()
df['MACD'] = ema_6 - ema_14
df['MACD_Signal'] = df['MACD'].ewm(span=4, adjust=False).mean()

# CMF (Chaikin Money Flow)
mf_multiplier = ((df['Close'] - df['Low']) - (df['High'] - df['Close'])) / (df['High'] - df['Low']).replace(0, np.nan)
mf_volume = mf_multiplier * df['Volume']
df['CMF'] = mf_volume.rolling(window=10).sum() / df['Volume'].rolling(window=10).sum()

# Display the technical indicators
technical_indicators = df[['Close', 'RSI_14', 'MACD', 'MACD_Signal', 'CMF']].copy()
technical_indicators
```

**Technical Indicators Explained:**

| **Indicator**     | **Description**                                                                              |
| ----------------- | -------------------------------------------------------------------------------------------- |
| **RSI\_14**       | Relative Strength Index (14-day): Measures momentum; <30 = oversold, >70 = overbought        |
| **MACD**          | MACD Line: Difference between 6-day and 14-day EMAs (momentum/trend)                        |
| **MACD\_Signal**  | 4-day EMA of MACD; used to identify crossovers (buy/sell signals)                            |
| **CMF\_10**       | Chaikin Money Flow (10-day): Combines price and volume to detect buying/selling pressure     |

### 4.1. ARIMA

#### 4.1.1. Stationarity using Augmented Dickey-Fuller (ADF) test

```python
from statsmodels.tsa.stattools import adfuller

# Use only the 'Close' column for modeling
data = df['Close']

# Apply Augmented Dickey-Fuller test
adf_result = adfuller(data)

# Collect the results
adf_output = {
    "ADF Statistic": adf_result[0],
    "p-value": adf_result[1],
    "Critical Values": adf_result[4]
}

adf_output
```

**ADF Test Result (on raw closing prices)**
- **ADF Statistic:** 3.63  
- **p-value:** 1.0  
- **Critical Value @ 5%:** -2.86  

**Interpretation:**  
Since:  
- The **ADF Statistic > critical value**  
- The **p-value = 1.0** (very high)  

â†’ We **fail to reject** the null hypothesis â€” the series is **not stationary**.

**Solution:** Apply **first-order differencing** (**d = 1**)


```python
# First-order differencing
differenced_series = data.diff().dropna()

# Re-run ADF test on differenced data
adf_diff_result = adfuller(differenced_series)

# Organize output
adf_diff_output = {
    "ADF Statistic": adf_diff_result[0],
    "p-value": adf_diff_result[1],
    "Critical Values": adf_diff_result[4]
}

adf_diff_output
```

**ADF Test After First-Order Differencing (d = 1)**
- **ADF Statistic:** -15.48  
- **p-value:** 2.6e-28  
- **Critical Value @ 5%:** -2.86  

**Interpretation:**  
- ADF Statistic is **much lower** than all critical values  
- p-value is **very small** (â‰ª 0.05)  

â†’ We **reject the null hypothesis** â†’ the differenced series is **stationary**.

**Conclusion:**  
- **First-order differencing is appropriate**  
- We should set **d = 1** in our ARIMA model


#### 4.1.2. Plotting ACF and PACF to find the best p and q

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(differenced_series, ax=axes[0], lags=40)
axes[0].set_title('Autocorrelation Function (ACF)')
plot_pacf(differenced_series, ax=axes[1], lags=40, method='ywm')
axes[1].set_title('Partial Autocorrelation Function (PACF)')
plt.tight_layout()
plt.show()
```

**ACF (Autocorrelation Function)** â†’ Helps Choose **q** (MA term)  
The ACF plot shows significant spike at lag **3**, then drops.

Suggests **q = 3** could be reasonable

**PACF (Partial Autocorrelation Function)** â†’ Helps Choose **p** (AR term)  
The PACF plot shows a significant spike at **lag 3**, then tapers off.

Suggests **p = 3**

**Suggested Model:** `ARIMA(p=3, d=1, q=3)`


```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Train/test split
train = data.iloc[:-30]
test = data.iloc[-30:]

# Fit ARIMA(3,1,3) model
model = ARIMA(train, order=(3, 1, 3))
model_fit = model.fit()

# Forecast next 30 steps
forecast = model_fit.forecast(steps=30)

# Evaluation
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

# Plot
plt.figure(figsize=(14, 6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, forecast, label='Forecast', color='red')
plt.title(f'ARIMA Forecast vs Actual\nMAE: {mae:.2f}, RMSE: {rmse:.2f}')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

(mae, rmse)
```

**Forecast Summary (Next 30 Business Days)**

- **MAE (Mean Absolute Error):** 130.46  
- **RMSE (Root Mean Squared Error):** 150.85  

These are the average prediction errors in USD. Lower values are better â€” this **ARIMA model** gives us a solid baseline.

### 4.4. LSTM (Long Short-Term Memory) Model


LSTM (**Long Short-Term Memory**) is a special kind of **Recurrent Neural Network (RNN)** designed to learn **sequential patterns** and **long-term dependencies** in time-series data (like stock prices).

**Why not use a regular neural network?**

A **typical neural network**:
- Sees all input at once (like a snapshot)  
- Has no memory of what happened before  

But **stock prices**:
- Depend on past values (**what happened in the last 7â€“30 days matters**)  
- Show patterns that unfold over time  

â†’ Thatâ€™s why we need **LSTMs** â€” they have **"memory cells"** that can remember patterns over time.


#### 4.4.1. Data Reshaping for LSTM

**LSTM expects a 3D tensor as input:**

shape = (samples, time_steps, features)

- **samples:** total training examples (rows)  
- **time_steps:** how many past days to look back (e.g., 7)  
- **features:** number of input features per day (e.g., Close, RSI, MACD)


```python
from sklearn.preprocessing import MinMaxScaler

# Scale features and target
min_max_scaler = MinMaxScaler()
scaled_features = min_max_scaler.fit_transform(features)
scaled_target = min_max_scaler.fit_transform(target.values.reshape(-1, 1))  # Reshapes the 1D array into a 2D column vector. LSTM requires 2D input for scaling: shape = (n_samples, n_features)

# Create sequences  (transforms time-series data into sequences so that the LSTM can learn from the past days (time steps) to predict the next day)
def create_sequences(X, y, time_steps=7):
    Xs, ys = [], []       # initialize lists to collect input/output pairs
    for i in range(len(X) - time_steps):
        Xs.append(X[i:i + time_steps])    # 7-day window of features
        ys.append(y[i + time_steps])      # target on day 8
    return np.array(Xs), np.array(ys)

time_steps = 7
X_lstm, y_lstm = create_sequences(scaled_features, scaled_target, time_steps)

# Train/test split (last 30 days for test)
X_train_lstm, X_test_lstm = X_lstm[:-30], X_lstm[-30:]
y_train_lstm, y_test_lstm = y_lstm[:-30], y_lstm[-30:]

# Output shapes
X_train_lstm.shape, y_train_lstm.shape, X_test_lstm.shape, y_test_lstm.shape
```

**LSTM input data is now ready:**

- **Training set:** 6,099 samples, with **7 time steps** and **20 features** per step  
- **Testing set:** 30 samples (for the last 30 days)


#### 4.4.2. Build and Train the LSTM model

##### 4.4.2.1. Simple LSTM model

```python
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import LSTM, Dense

from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.losses import MeanSquaredError

# Build a simple LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, activation='relu', input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]))) # Adds an LSTM layer to the model, the shape of each input sample: a 7-day sequence with 20 features per day.
model_lstm.add(Dense(1))  # fully connected output layer for single-step regression (a single prediction: the stock price for the next day)

# Compile the model (prepares the model for training by specifying Update weights and Measure error methods)
model_lstm.compile(optimizer='adam', loss=MeanSquaredError())

# Train the model
history = model_lstm.fit(
    X_train_lstm, y_train_lstm,
    epochs=20,
    batch_size=32,
    validation_data=(X_test_lstm, y_test_lstm),
    verbose=1
)

# Plot Training History
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training & Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Forecast and Inverse Scale
y_pred_lstm = model_lstm.predict(X_test_lstm)
y_pred_inv = min_max_scaler.inverse_transform(y_pred_lstm)
y_test_inv = min_max_scaler.inverse_transform(y_test_lstm)

# Plot Actual vs Forecast
test_dates = df.index[-30:]
plt.figure(figsize=(12, 5))
plt.plot(test_dates, y_test_inv, label='Actual', marker='o')
plt.plot(test_dates, y_pred_inv, label='LSTM Forecast', marker='x')
plt.title('LSTM Forecast vs Actual (Last 30 Days)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

# Compute MAE and RMSE
mae = mean_absolute_error(y_test_inv, y_pred_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))

mae, rmse
```

**A simple LSTM model has been trained and tested.**  
Here's how it performed on the last 30 days of Amazon stock prices:

- **MAE (Mean Absolute Error):** $45.85

- **RMSE (Root Mean Squared Error):** $59.87  

##### 4.4.2.2. Deeper LSTM model (2 layers)

```python
from keras.layers import LSTM, Dense, Dropout

# Build a deeper LSTM model
model = Sequential()

# First LSTM layer (returns sequences for stacking)
model.add(LSTM(units=64, activation='relu', return_sequences=True, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model.add(Dropout(0.2))  # Dropout helps reduce overfitting

# Second LSTM layer (no return_sequences needed here)
model.add(LSTM(units=32, activation='relu'))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(1))  # Predicting one value (next-day Close)

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(
    X_train_lstm, y_train_lstm,
    epochs=30,
    batch_size=32,
    validation_data=(X_test_lstm, y_test_lstm),
    verbose=1
)

# Plot Training History
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training & Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Forecast and Inverse Scale
y_pred_lstm = model.predict(X_test_lstm)
y_pred_inv = min_max_scaler.inverse_transform(y_pred_lstm)
y_test_inv = min_max_scaler.inverse_transform(y_test_lstm)

# Plot Actual vs Forecast
test_dates = df.index[-30:]
plt.figure(figsize=(12, 5))
plt.plot(test_dates, y_test_inv, label='Actual', marker='o')
plt.plot(test_dates, y_pred_inv, label='LSTM Forecast', marker='x')
plt.title('LSTM Forecast vs Actual (Last 30 Days)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

# Compute MAE and RMSE
mae = mean_absolute_error(y_test_inv, y_pred_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))

mae, rmse
```

**A deep (2 layers) LSTM model has been trained and tested.**

##### 4.4.2.3. Optimized 2-layers LSTM model

A revised **2-layers LSTM model** with:

- **Reduced complexity** (fewer units: 32 â†’ 16)  
- **tanh activation** (better for LSTM temporal learning)  
- **EarlyStopping** (stop training when validation loss stops improving)  
- **More epochs** (100 max, with early stopping to exit early)


```python

from keras.callbacks import EarlyStopping

# Build a reduced-complexity LSTM model
model = Sequential()

# First LSTM layer
model.add(LSTM(units=32, activation='tanh', return_sequences=True, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model.add(Dropout(0.2))

# Second LSTM layer
model.add(LSTM(units=16, activation='tanh'))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(1))  # Predict one value (next-day Close)

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Setup EarlyStopping    (automatically stop training when the model stops improving on the validation set)
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history = model.fit(
    X_train_lstm, y_train_lstm,
    epochs=100,
    batch_size=32,
    validation_data=(X_test_lstm, y_test_lstm),
    callbacks=[early_stop],
    verbose=1
)

# Plot Training History
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training & Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Forecast and Inverse Scale
y_pred_lstm = model.predict(X_test_lstm)
y_pred_inv = min_max_scaler.inverse_transform(y_pred_lstm)
y_test_inv = min_max_scaler.inverse_transform(y_test_lstm)

# Plot Actual vs Forecast
test_dates = df.index[-30:]
plt.figure(figsize=(12, 5))
plt.plot(test_dates, y_test_inv, label='Actual', marker='o')
plt.plot(test_dates, y_pred_inv, label='LSTM Forecast', marker='x')
plt.title('LSTM Forecast vs Actual (Last 30 Days)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

# Compute MAE and RMSE
mae = mean_absolute_error(y_test_inv, y_pred_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))

mae, rmse
```

**An optimized 2-layers LSTM model has been trained and tested.**  
Here's how it performed on the last 30 days of Amazon stock prices:

- **MAE (Mean Absolute Error):** $54.32

- **RMSE (Root Mean Squared Error):** $70.10  

##### 4.4.2.4. Bi-directional LSTM model (BiLSTM)

A **Bidirectional LSTM** processes each sequence twice:

- **Forward pass:** like a normal LSTM (**t=1 â†’ t=7**)  
- **Backward pass:** runs in reverse (**t=7 â†’ t=1**)  

Then, it **combines both outputs** at each time step.


```python
from keras.layers import LSTM, Dense, Dropout, Bidirectional
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Build Bidirectional LSTM model
model = Sequential()

# First BiLSTM layer
model.add(Bidirectional(LSTM(units=32, activation='tanh', return_sequences=True), input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model.add(Dropout(0.2))

# Second BiLSTM layer
model.add(Bidirectional(LSTM(units=16, activation='tanh')))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(1))  # One value per forecast

# Compile
model.compile(optimizer='adam', loss='mse')

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
lr_schedule = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5)      # learning rate scheduler automatically reduces the learning rate during training if the model stops improving

# Train the model
history = model.fit(
    X_train_lstm, y_train_lstm,
    epochs=100,
    batch_size=32,
    validation_data=(X_test_lstm, y_test_lstm),
    callbacks=[early_stop, lr_schedule],
    verbose=1
)

# Plot Training History
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training & Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Forecast and Inverse Scale
y_pred_lstm = model.predict(X_test_lstm)
y_pred_inv = min_max_scaler.inverse_transform(y_pred_lstm)
y_test_inv = min_max_scaler.inverse_transform(y_test_lstm)

# Plot Actual vs Forecast
test_dates = df.index[-30:]
plt.figure(figsize=(12, 5))
plt.plot(test_dates, y_test_inv, label='Actual', marker='o')
plt.plot(test_dates, y_pred_inv, label='LSTM Forecast', marker='x')
plt.title('BiLSTM Forecast vs Actual (Last 30 Days)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

# Compute MAE and RMSE
mae = mean_absolute_error(y_test_inv, y_pred_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))

mae, rmse
```

**A BiLSTM model has been trained and tested.**  
Here's how it performed on the last 30 days of Amazon stock prices:

- **MAE (Mean Absolute Error):** $55.03

- **RMSE (Root Mean Squared Error):** $69.23  

**The simple LSTM model commonly perform better than deeper one, especially in certain situations when the deep model has many limitations like: Overfitting, Too Complex for the Data,Training Instability, Insufficient Training Time and Bad Weight Initialization or Optimizer Dynamics.**
